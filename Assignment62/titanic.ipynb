```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,Dropout,BatchNormalization
```

```python
data_train = pd.read_csv('/content/train.csv')

data_train
```

```python
data_test = pd.read_csv("/content/test.csv")
data_test
```

```python
data_train.replace(['female','male'],[0,1],inplace=True)

data_train.fillna(20,inplace=True)
```

```python
data_test.replace(['female','male'],[0,1],inplace=True)

data_test.fillna(20,inplace=True)
```

```python
X_train = np.array(data_train[['Pclass','Sex','Age','SibSp','Fare']])
Y_train = np.array(data_train[['Survived']])

```

```python
X_test = np.array(data_test[['Pclass','Sex','Age','SibSp','Fare']])
```

```python
Y_test = pd.read_csv("/content/gender_submission.csv")
Y_test = Y_test[["Survived"]].to_numpy()
```

```python

train_dataset = tf.data.Dataset.from_tensor_slices((X_train,Y_train)).batch(32).shuffle(1000)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test,Y_test)).batch(32)
```

```python
class MyModel(tf.keras.Model):
  def __init__(self):
    super().__init__()
    # self.conv1 = Conv2D(64,(2,2), activation="relu", padding="same")
    # self.conv2 = Conv2D(64, (4,4), activation="relu", padding="same")
    # self.conv3 = Conv2D(128, (2,2), activation="relu", padding="same")
    # self.conv4 = Conv2D(32, (3,3), activation="relu", padding="same")
    # self.b1 = BatchNormalization()
    # self.b2 = BatchNormalization()
    # self.b3 = BatchNormalization()
    # self.b4 = BatchNormalization()
    # self.m = MaxPooling2D(3,3)
    # self.dropout1 = Dropout(0.3)
    # self.dropout2 = Dropout(0.5)
    # self.dropout3 = Dropout(0.4)
    self.dense1 = Dense(5, activation='sigmoid')
    self.dense2 = Dense(18, activation='relu')
    self.dense3 = Dense(10, activation='softmax')
    self.dense4 = Dense(2, activation='softmax')
    #self.dense2 = Dense(10, activation='softmax')
    #self.f = Flatten()
  def call(self, x):
    x = self.dense1(x)
    x = self.dense2(x)
    x = self.dense3(x)
    x = self.dense4(x)
    return x

model = MyModel()
```

```python
# Main loss function
loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)


# Metrics
train_loss = tf.keras.metrics.Mean()
test_loss = tf.keras.metrics.Mean()

train_acc = tf.keras.metrics.SparseCategoricalAccuracy()
test_acc = tf.keras.metrics.SparseCategoricalAccuracy()
```

```python

@tf.function
def train(sample_data , label):
    with tf.GradientTape() as tape :
      prediction = model(sample_data)
      loss = loss_function(label , prediction)
    gradients = tape.gradient(loss , model.trainable_variables)
    optimizer.apply_gradients(grads_and_vars= zip(gradients , model.trainable_variables))

    train_loss(loss)
    train_acc(label , prediction)
```

```python
@tf.function
def test(sample_data , label):
  prediction = model(sample_data)
  loss = loss_function(y_true=label , y_pred=prediction)
  test_loss(loss )
  test_acc(label , prediction)
```

```python
epochs = 100
for epoch in range(epochs):
  train_loss.reset_states()
  train_acc.reset_states()
  test_loss.reset_states()
  test_acc.reset_states()

  #train
  for data , label in train_dataset :
    train(data , label)

  # test
  for data , label in test_dataset :
    test(data , label)

  print("epoch:" , epoch + 1  ,
        f"Train Loss: {train_loss.result()}" ,
        f"Train Acc: {train_acc.result()}" ,
        f"Test Loss: {test_loss.result()}" ,
        f"Test Acc: {test_acc.result()}"
        )
```
