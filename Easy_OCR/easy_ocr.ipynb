```python
import easyocr
from PIL import Image
```

```python
import PIL.Image
if not hasattr(PIL.Image, 'Resampling'):  
    PIL.Image.Resampling = PIL.Image
```

## Latin Infrence

```python
im = Image.open('Inference\Latin\handwritten.jpg')
im2 = im.resize((256,256),resample=Image.BICUBIC)
display(im2)
```

```python
reader = easyocr.Reader(['en']) 
en_result = reader.readtext('Inference\Latin\handwritten.jpg')
en_result
```

```python
im = Image.open('Inference\Latin\license_plate.jpg')
im2 = im.resize((256,256),resample=Image.BICUBIC)
display(im2)
```

```python
reader = easyocr.Reader(['en']) 
en_result = reader.readtext('Inference\Latin\license_plate.jpg')
en_result
```
